```python
def transcription(file):
    with open(file, "rb") as audio_file:
        transcript = client.audio.transcriptions.create(
            model = "whisper-1",
            file = audio_file
        )

    return transcript.text
```

```python
base = transcription("/Users/Rakshit.Lodha/evals/Base.wav")
```

```python
hindi = transcription("/Users/Rakshit.Lodha/evals/Hindi.wav")
```

```python
john = transcription("/Users/Rakshit.Lodha/evals/John.wav")
```

```python
transcripts = [
    "Hey, hi, I'm Rakshit and I work as a product manager at a company called ET Money",
    "I'm John and I belong to the Unites States of America and I work as a product manager at a company called ET Money.",
    "नमस्ते मेरा नाम रक्षित है और मैं एक कंपनी में काम करता हूं जिसका नाम ET Money है।"
    ]
```

```python
wer_score = wer(base.lower(), transcripts[0].lower())

print(base)
print(transcripts[0])

print(wer_score)

cer_score = cer(base.lower(), transcripts[0].lower())

print(cer_score)
```
